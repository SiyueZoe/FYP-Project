# Sensitivity Test

### Overview: 
1. Test Case 1.1 to 1.5: How Pmin will affect power distribution?
2. Summary of Test Case 1.1 to 1.5: How Pmin will affect total cost?
3. Test Case 2.1 to 2.5: How Pmin will affect power distribution? (with another set of parameters)

5. Test Case 3.1 and 3.2: Comparison between IL w/ EM and IL w/o EM when Pmin is extremely small
6. Test Case 4.1 and 4.2: Comparison between IL w/ EM and IL w/o EM when Pmin is extremely small (with another set of parameters)

```python
import numpy as np
from IL_with_EM_MILP import IL_MILP
from IL_LP import IL_LP

import matplotlib.pyplot as plt
```

## Test Case 1.1

```python
# Pmin = 10% * P
Total_Cost = []
dt = 10
pr = np.array([1,50,1,10,10])
P = 20
E = 500
T_off = 2
Pmin = 0.1 * P
N = len(pr)

# MILP Solution
solution = IL_MILP(dt, pr, P, E, T_off, Pmin)
Total_Cost.append((solution[5 * N]))
print('Pmin = ', Pmin)

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[0:N], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 
1) Power status are all correct 
2) At the 2nd time step, price is high and thus power is equal to the Pmin to ensure ON status to avoid frequent switch. 
3) All requirements are satisfied.
