# Sensitivity Test

```python
import numpy as np
from IL_with_EM_MILP import IL_MILP

import matplotlib.pyplot as plt
```

## Test Case 1.1

```python
# Pmin = 10% * P
Total_Cost = []
dt = 10
pr = np.array([1,50,1,10,10])
P = 20
E = 500
T_off = 2
Pmin = 0.1 * P
N = len(pr)

# MILP Solution
solution = IL_MILP(dt, pr, P, E, T_off, Pmin)
Total_Cost.append((solution[5 * N]))
# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[0:N], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 
1) Power status are all correct 
2) At the 2nd time step, price is high and thus power is equal to the Pmin to ensure ON status to avoid frequent switch. 
3) All requirements are satisfied.
