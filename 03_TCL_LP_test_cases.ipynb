# Thermostatically Controlled Load using Linear Programming Testing

```python
import numpy as np
from TCL_LP import TCL_LP

import matplotlib.pyplot as plt
```

## Test Case 1

```python
pr = np.array([10, 20, 30, 40, 50])
dt = 10
P = 20
c_water = 0.073
m = 50
temp_up = 25
temp_o = 25
temp_req = 37
temp_en = np.array([25, 20, 18, 15, 15])
di = np.array([25, 27, 28, 22, 25])

# LP Solution
solution = TCL_LP(dt, pr, P, c_water, m, temp_up, temp_o, temp_req, temp_en, di)

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[1], 'x-')
ax.plot(solution[2], 'v-')
ax.plot(solution[3], 'p-')
ax.plot(solution[4], 'm-')
ax.text(4.5,0,'power status',color='orange')
ax.text(4.5,10,'price',color='blue')
ax.text(4.5,20,'lower bound',color='red')
ax.text(4.5,30,'optimal energy',color='purple')
ax.text(4.5,40,'upper bound',color='green')
ax.plot()


ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 1) power at each status is within limit. 2) Energy at each time step satisfies constraints.(This is the special case as initial temperature and upper limit value are set the same, so that the upper bound and lower bound of energy are the same and then there is only one optimal value)
