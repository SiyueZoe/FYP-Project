# All Loads with Energy Limit using MILP Testing

```python
import numpy as np
from All_with_Energy_Limit_MILP import ALL_with_Energy_Limit_MILP

import matplotlib.pyplot as plt
```

## Test Case 1

```python
# Parameter Setting
pr = np.array([10, 50, 10, 50, 10])
dt = 10
E_limit = 1500
# For NL
L = 3
P_NL = 20
# For IL
P_IL = 20
E_IL = 500
T_off = 3
Pmin = 0.4 * P_IL
# For TCL
P_TCL = 10
c_water = 0.073
m = 50
temp_up = 25
temp_o = 20
temp_req = 60
temp_en = np.array([25, 20, 18, 15, 15])
di = np.array([25, 27, 28, 22, 25])

# MILP Solution
solution = ALL_with_Energy_Limit_MILP(dt, pr, E_limit, L, P_NL, P_IL, E_IL, T_off, Pmin, P_TCL, c_water, m, temp_up, temp_o,
                                  temp_req, temp_en, di)
# Plot
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[0][:], 'x-')
ax.plot(solution[2][:], 'v-')
ax.plot(solution[3][:], 'm-')
ax.plot(solution[1][:], 'p-')
ax.text(4.5,0,'power status',color='orange')
ax.text(4.5,20,'price',color='blue')
ax.text(4.5,40,'lower bound',color='red')
ax.text(4.5,60,'optimal energy',color='purple')
ax.text(4.5,80,'upper bound',color='green')
ax.plot()
ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: The scheduling problem consists of three types of loads: NL, IL, TCL. For NL, the needed power value and duration of that is fixed and thus the energy consumed by NL is fixed. For IL, there is one constraint to ensure the total energy for IL is a certain value and hence the energy consumed by TCL is also fixed. For TCL, there is one upper energy bound and one lower energy bound. Without introducing energy limit parameter, the total energy consumed by TCL is always equal to the lower energy bound, which means the energy for TCL is also fixed. When energy consumption for each type of load is fixed, the energy limit parameter is meaningless to be added in. If energy limit is added in, when it is too large, it has no use; when it is too small, there is no result shown.
