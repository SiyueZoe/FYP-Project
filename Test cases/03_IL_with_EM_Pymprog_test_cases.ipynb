# Interruptible Load with EM using Pymprog Testing

```python
import numpy as np
from IL_Pymprog import IL_Pymprog

import matplotlib.pyplot as plt
```

## Test Case 1

```python
dt = 10
pr = np.array([10,20,30,40,50,60])
P = 20
E = 900
T_off = 4
small_const = 0.0001
N = len(pr)

# MILP solution
solution = IL_Pymprog(dt, pr, P, E, T_off, small_const)

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[0:N], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 1) The sum of energy is correct. 2) This follows the connection between x and y. 3) Ancillary binary y follows the requirement of switching on/off.
