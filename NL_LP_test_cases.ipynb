# Non-interruptible Load with Linear Programming Testing

```python
import numpy as np
from NL_LP import NL_LP

import matplotlib.pyplot as plt
```

## Test Case 1

```python
dt = 10
pr = np.array([50, 10, 50, 10, 50])
L = 2
P = 20

# LP solution 
solution = NL_LP(dt, pr, L, P)
print(solution)

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[1][:], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 1) The sum of power is correct. 2) This violates the fact that power can be either 0 or P. 3) This violates non-interruptability for L time steps
