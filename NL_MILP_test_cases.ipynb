# Non-interruptible Load with Mixed-integer Linear Programming Testing

```python
import numpy as np
from NL_MILP import NL_MILP

import matplotlib.pyplot as plt
```

## Test Case 1

```python
dt = 10
pr = [50, 10, 50, 10, 50]
L = 2
P = 20
N = len(pr)

# MILP solution 
solution = NL_MILP(dt, pr, L, P)

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[0:N], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Comments: 1) The sum of power is correct. 2) This follows the fact that power can be either 0 or P. 3) This follows non-interruptability for L time steps
