# IL LP with EM module testing
This file includes test cases for the LP relaxation of Interruptable Load scheduling with minimum time off constraints.

```python
import numpy as np
from IL_LP_with_EM import IL_LP

import matplotlib.pyplot as plt
```

# Test case 1

The monotonic price increase

Expectation: the continous consumption through first time intervals

```python
dt = 10
pr = np.array([10,20,30,40,50,60])
P = 20
E = 900
T_off = 4

# LP solution
solution = IL_LP(dt, pr, P, E, T_off)
print('On/Off statuses:', solution[2][:])

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[1][:], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Result of the test case 1: On/Off status on the 4th interval is non-discrete because of the LP relaxation. Constraints are not violated. The IL_LP module works correctly. The optimal solution does not coincide with MILP (Mixed Integer Linear Programming) in terms of ancillary binary variables 'y' and coincides in terms of real power consumption 'x'

# Test case 4

The smallest price at the beginning and then the price peak

Expectation: Consumption should be very low during the price peak

```python
dt = 10
pr = np.array([1,50,1,10,10])
P = 20
E = 500
T_off = 2

# LP solution
solution = IL_LP(dt, pr, P, E, T_off)
print('On/Off statuses:', solution[2][:])

# Plot the price distribution
fig, ax = plt.subplots()
ax.plot(pr, 'o-')
ax.plot(solution[1][:], 'x-')
ax.plot()

ax.set(xlabel='Time', ylabel='Price, $ (dots)/ \n Consumption, Wh (crosses)',
       title='Prices and Load Scheduling');
```

Result of the test case 4: The last On\Off statuses are non-discrete. Constraints are violated as $x_i = y_iP \; \forall i$ but power status y for time 1 is 'On' while the real consumption x is 0. The IL_LP module works incorrectly and the optimal solution does not coincide with MILP (Mixed Integer Linear Programming) in terms of real power consumption 'x'
